# 4.There is Outliers Present in the data soo for removing the Outliers i use below steps

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
```

```python
df=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
```

# Using Machine Learning Algorithms

```python
#Splitting Data into Train & Test
```

```python
x=df.drop(['Price'],axis=1)

y=df['Price']              
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.3, random_state = 1)
```

# Training the Model : Linear Regression

```python
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score
 
model1=LinearRegression()# get instance of model

model1.fit(xtrain,ytrain)# Train/Fit model

ypred1=model1.predict(xtest) # get y predictions

r2=r2_score(ytest,ypred1)

print('The accuracy od Linear Regression is :',r2*100)
```

```python
train=model1.score(xtrain,ytrain)
test=model1.score(xtest,ytest)
print(train)
print(test)
```

# Training the Model : RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor()# get instance of model

model2.fit(xtrain,ytrain) # Train/Fit model

ypred2 = model2.predict(xtest) # get y predictions


print('The accuracy of the Random Forest is', model2.score (xtest, ytest)*100)
```

```python
train=model2.score(xtrain,ytrain)
test=model2.score(xtest,ytest)
print(train)
print(test)
```

# Training the Model : Model: KNN (K-Nearest Neighbours)

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsRegressor

model3 = KNeighborsRegressor(n_neighbors=5) # get instance of model

model3.fit(xtrain, ytrain) # Train/Fit model

ypred3 = model3.predict(xtest) # get y predictions

print('The accuracy of the K-Nearest Neighbours is', model3.score(xtest, ytest)*100)
```

```python
train=model3.score(xtrain,ytrain)
test=model3.score(xtest,ytest)
print(train)
print(test)
```

# Tunning of KNN model

```python
trainac=[]
testac=[]

for i in range(35,250):
    model3=KNeighborsRegressor(n_neighbors=i)
    model3.fit(xtrain,ytrain)
    train=model3.score(xtrain,ytrain)
    test=model3.score(xtest,ytest)
    trainac.append(train)
    testac.append(test)
```

```python
plt.plot(range(35,250),trainac,color="red")
plt.plot(range(35,250),testac,color="green")
plt.grid()
plt.show()
```

```python
model3 = KNeighborsRegressor(n_neighbors=45) # get instance of model

model3.fit(xtrain, ytrain) # Train/Fit model

ypred3 = model3.predict(xtest) # get y predictions

print('The accuracy of the K-Nearest Neighbours is', model3.score(xtest, ytest)*100)
```

# Training the Model : DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
model4=DecisionTreeRegressor()
model4.fit(xtrain,ytrain)
ypred4=model4.predict(xtest)
ypred4
model4.score(xtest,ytest)*100
```

```python
train=model4.score(xtrain,ytrain)
test=model4.score(xtest,ytest)
print(train)
print(test)
```

# We got highest Accuracy  in RandomForestRegressor Model , now we try maximize the accuracy by Hyperparameter RandomForestRegressor

```python
from sklearn.model_selection import GridSearchCV

model2 = RandomForestRegressor()
# Create a search grid of parameters that will be shuffled through

param_grid = {


'n_estimators': [100,200, 300],    
'max_depth': [None,4,8],
'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(model2,param_grid,cv=5,
                          scoring='neg_mean_squared_error',
                          return_train_score=True)
grid_search.fit(xtrain,ytrain)
```

```python

grid_search.best_estimator_
```
